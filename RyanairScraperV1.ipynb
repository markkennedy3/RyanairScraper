```python
from time import sleep, strftime
from random import randint
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import smtplib
from email.mime.multipart import MIMEMultipart
from selenium.common.exceptions import NoSuchElementException
import numpy as np
```

```python
chromedriver_path = '/Users/markkennedy/Documents/WebDriver/chromedriver' # Change this to your own chromedriver path
```

```python
driver = webdriver.Chrome(executable_path=chromedriver_path)
sleep(5)
```

```python
ryanair = 'https://www.ryanair.com/ie/en/cheap-flights'
driver.get(ryanair)
sleep(10)
```

```python
terms_and_conditions = '//span[contains(@class,"terms-conditions-checkbox-span")]'
driver.find_elements_by_xpath(terms_and_conditions)[1].click()
```

```python
lets_go = '//button[contains(@class,"core-btn-primary-i core-btn-block core-btn-big")]'
driver.find_element_by_xpath(lets_go).click()
sleep(3)
```

```python
sections_list = []
```

```python
def page_scrape():
    xp_sections = '//*[@class="ff-list-item"]'
    sections = driver.find_elements_by_xpath(xp_sections)
    temp_sections_list = ([value.text for value in sections])
    temp_sections_list = [x.replace('in ', "") for x in temp_sections_list]
    temp_sections_list = ([value.split("\n") for value in temp_sections_list])
    sections_list.append(temp_sections_list)
```

```python
def load_next_page():
    next_page = '//div[contains(@class,"next vertical-center")]'
    driver.find_elements_by_xpath(next_page)[0].click()
    sleep(3)
```

```python
def check_next_page():
    next_page = '//a [contains(@class,"core-link disabled")]'
    contents = driver.find_elements_by_xpath(next_page)
    if contents == []:
        return True
    else:
        return False
```

```python
page_scrape()

next_page = '//div[contains(@class,"next vertical-center")]'
driver.find_elements_by_xpath(next_page)[0].click()

sleep(2)
```

```python
while True:
        page_scrape()
        load_next_page()
        if check_next_page() == False:
            break
```

```python
page_scrape()
```

```python
sections_array = np.concatenate(sections_list, axis=0 )
```

```python
columns = ["City", "Country", "Month", "Price", "Ticket"]
df = pd.DataFrame(data=sections_array, columns=columns)
```
